```
# all images are load in variable x...
FILE='TesteSugarAbr25.zip'
img_name=[]
x=[]
with zipfile.ZipFile(FILE, "r") as f:
    for name in f.namelist():
        img_name.append(name)
        x.append(cv2.imread(name))

plt.axis("off")
plt.imshow(x[0])
```
