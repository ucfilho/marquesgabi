```
%matplotlib inline
from matplotlib import pyplot as plt
import os 
import cv2
import zipfile
```

```
file_name = zipfile.ZipFile('TesteSugarAbr25.zip', 'r')
file_name.extractall()
```

```
# all images are load in variable x...
FILE='TesteSugarAbr25.zip'
img_name=[]
x=[]
with zipfile.ZipFile(FILE, "r") as f:
    for name in f.namelist():
        img_name.append(name)
        x.append(cv2.imread(name))

plt.axis("off")
plt.imshow(x[0])
```

```
laplacian=cv2.Laplacian(x[0],cv2.CV_64F)
plt.axis("off")
plt.imshow(laplacian)
```

```
gray = cv2.cvtColor(x[0], cv2.COLOR_BGR2GRAY)
plt.axis("off")
plt.imshow(gray)
```

```
destRGB = cv2.cvtColor(x[0], cv2.COLOR_RGB2BGR)

gray_02 = cv2.cvtColor(destRGB, cv2.COLOR_RGB2GRAY)
plt.axis("off")
plt.imshow(gray_02)
```
