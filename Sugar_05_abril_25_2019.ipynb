```

file_name = zipfile.ZipFile('TesteSugarAbr25.zip', 'r')
file_name.extractall() 
```

```
# all images are load in variable x...
FILE='TesteSugarAbr25.zip'
img_name=[]
x=[]
with zipfile.ZipFile(FILE, "r") as f:
    for name in f.namelist():
        img_name.append(name)
        x.append(cv2.cvtColor(cv2.imread(name), cv2.COLOR_BGR2RGB))

plt.axis("off")
plt.imshow(x[0])
```

```
resized_image = cv2.resize(x[0], (2,2))
for i in resized_image:
  print(i)
```

```
Indice=x[0].shape
print(Indice)
print(x[0][0,0,0])
print(Indice[0])
```

```
N1=Indice[0]
N2=Indice[1]
N3=Indice[2]
Ref=200
for i in range(N1):
  for j in range(N2):
    for k in range(N3):
      if(x[0][i,j,k]<Ref):
        x[0][i,j,k]=0
```

```
plt.axis("off")
plt.imshow(x[0])
```

```
Indice=x[1].shape
N1=Indice[0]
N2=Indice[1]
N3=Indice[2]
Ref=500
for i in range(N1):
  for j in range(N2):
    for k in range(N3):
      if(sum(x[1][i,j,:])>Ref):
        x[1][i,j,k]=255
      else:
        x[1][i,j,k]=0
```

```
plt.axis("off")
plt.imshow(x[1])
```
